```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

```python
df = pd.read_csv('satellite_sensor_bands.csv')
```

```python
fig, ax = plt.subplots(figsize=(15,5))

# Set up colors
categories = np.unique(df.Sensor)
colors = np.linspace(0, 1, len(categories))
colordict = dict(zip(categories, colors)) 
df['color'] = df.Sensor.apply(lambda x: colordict[x])
cmap = cm.get_cmap('tab20')

# Plot
ax.hlines(df.Sensor, df.BandwidthLow, df.BandwidthHigh, linewidth=15, color=cmap(df.color))
ax.grid(b=True, which='major', axis='x', linestyle='-')
ax.grid(b=True, which='minor', axis='y', linestyle=':')
ax.set_xlim((0,15))
ax.set_xlabel('Wavelength [$\mu m$]')
ax.set_title('Satellite Sensor Band Wavelengths');
```
